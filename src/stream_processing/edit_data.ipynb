```python
def add_id(df, year, month):
    list_id = []
    for i in range(len(df)):
        _id = year.__str__() + month.__str__() + (i+1).__str__().zfill(8)
        list_id.append(_id)
    df["id"] = list_id
    return df
```

```python
def add_id_customer(df, list_id):
    list_id_cus = []
    for i in range(len(df)):
        ran_idx = random.randint(0, len(list_id)-1)
        list_id_cus.append(list_id[ran_idx])
    df["id_customer"] = list_id_cus
    return df
```

```python
def add_rate(df):
    list_rate = []
    for i in range(len(df)):
        rate = round(random.random() * 5, 1) + 1
        if rate > 5:
            rate = 5
        list_rate.append(rate)
    df["rate"] = list_rate    
    return df
```

```python
def split_follow_day(df):
    list_follow_day = []
    start_date = df['lpep_dropoff_datetime'].min().date()
    end_date = df['lpep_dropoff_datetime'].max().date()
    current_date = start_date
    daily_dfs = {}
    while current_date <= end_date:
        # Lọc dữ liệu cho ngày hiện tại
        mask = df['lpep_dropoff_datetime'].dt.date == current_date
        daily_df = df[mask]
        
        # Lưu vào dict với key là ngày
        if not daily_df.empty:
            daily_dfs[current_date] = daily_df
        
        # Chuyển sang ngày tiếp theo
        current_date += timedelta(days=1)
    
    return daily_dfs
    
        
```
