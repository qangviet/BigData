```python
from kafka import KafkaConsumer
import json
```

```python
consumer = KafkaConsumer(
    'streaming.public.green_trip_raw_2024_4',
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))
)
```

```python
for message in consumer:
    
    payload = message.value

    # Check if 'after' field exists
    payload = message.value

    # Check if 'payload' and 'after' fields exists
    if 'payload' in payload and 'after' in payload['payload']:
        after_data = payload['payload']['after']

        # In hoặc xử lý dữ liệu 'after'
        print("Data After:", after_data)

        # Thêm logic xử lý dữ liệu ở đây (ví dụ: ghi vào database, xử lý tiếp, etc.)
    else:
        print("No 'payload' or 'after' field found in message:", payload)
    
    print(message)
```
